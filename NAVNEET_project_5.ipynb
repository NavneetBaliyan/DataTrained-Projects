```python
from scipy.stats import zscore
zscore=abs(zscore(df_wine))
print(df_wine.shape)
df_wine1=df_wine.loc[(zscore<3).all(axis=1)]
print(df_wine1.shape)
```

```python
import warnings
warnings.filterwarnings("ignore")
for col in df_wine1.columns:
    if df_wine1.skew().loc[col]>0.55:
        df_wine1[col]=np.log1p(df_wine1[col])
```

```python
x=df_wine1.iloc[:,0:10]
```

```python
y=df_wine1.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=35,test_size=0.02)
                                              
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import  DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import accuracy_score
model= [LogisticRegression(), GaussianNB(), SVC(), DecisionTreeClassifier(), KNeighborsClassifier(),
       MultinomialNB() , RandomForestClassifier()]
```

```python
import warnings
warnings.filterwarnings("ignore")
for m in model:
    
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('accuracy score of m',m,'is')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
```

```python
#use AdaBoostClassifier to improve the accuracy of model
from sklearn.ensemble import AdaBoostClassifier
abc=AdaBoostClassifier(n_estimators=30,base_estimator=DecisionTreeClassifier(),learning_rate=2)
m=abc.fit(x_train,y_train)
predm=m.predict(x_test)
```

```python
from sklearn import metrics
print("accuracy_score",metrics.accuracy_score(y_test,predm))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(DecisionTreeClassifier(),x,y,cv=5,scoring="r2").mean()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
print("r2_score is :",r2_score(y_test,predm))
print("RMSE is :",np.sqrt(mean_squared_error(y_test,predm)))
```
